```python
# here the image is loaded using OpenCV
im = cv2.imread(image_path)
imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
```

```python
# here the image is loaded using OpenCV
hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
h_hist = cv2.calcHist([hsv], [0], None, [180], [0,180])
s_hist = cv2.calcHist([hsv], [1], None, [256], [0,256])
v_hist = cv2.calcHist([hsv], [2], None, [256], [0,256])
```

```python
plot(h_hist)
```

```python
dom_h = uint8(argmax(h_hist))*ones(100)
dom_s = uint8(argmax(s_hist))*ones(100)
dom_v = uint8(argmax(v_hist))*ones(100)
dom_hsv = zeros((100,100,3))
dom_hsv[:,:,0] = dom_h
dom_hsv[:,:,1] = dom_s
dom_hsv[:,:,2] = dom_v
dom_hsv = array(dom_hsv, dtype = uint8)
```

```python
dom_hsv[:,:,0]
```

```python
dom_rgb = cv2.cvtColor(dom_hsv, cv2.COLOR_HSV2RGB)
```

```python
imshow(dom_rgb)
```

# Dominant RGB

```python
b_hist = cv2.calcHist([im],[0],None,[256],[0,256])
g_hist = cv2.calcHist([im],[1],None,[256],[0,256])
r_hist = cv2.calcHist([im],[2],None,[256],[0,256])
```

```python
dom_b = argmax(b_hist)*ones(100)
dom_g = argmax(g_hist)*ones(100)
dom_r = argmax(r_hist)*ones(100)
dom_rgb = zeros((100,100,3))
dom_rgb[:,:,0] = dom_r
dom_rgb[:,:,1] = dom_g
dom_rgb[:,:,2] = dom_b
```

```python
imshow(dom_rgb)
```
